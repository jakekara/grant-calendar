```python
import pandas as pd
import re
```

```python
def clean_str(s):
    try:
        return re.sub(r"[^a-zA-Z0-9 ]+","",s).strip()
    except:
        return s
```

```python
grant_schedule = pd.read_csv("https://docs.google.com/spreadsheets/d/1AEALzh1Gwx_WzQnJk4bq_rrsv_SwFgqZxlpl_imgqkk/pub?gid=0&single=true&output=csv")
grant_schedule["Town Grant Name"] = grant_schedule["Town Grant Name"].apply(clean_str)
grant_schedule.to_csv("../data/calendar.csv",index=False)
grant_schedule.head()
```

```python
town_runs = pd.read_csv("https://docs.google.com/spreadsheets/d/1AEALzh1Gwx_WzQnJk4bq_rrsv_SwFgqZxlpl_imgqkk/pub?gid=825246286&single=true&output=csv",
                        header=[0,1])

keepers = filter(lambda x: "Unnamed:" not in x[0], town_runs.columns)

town_runs = town_runs[keepers]
town_runs.columns = map(lambda x: clean_str(x[0]) + "_" + clean_str(x[1]), keepers)
town_runs = town_runs.set_index("Town_Town")
town_runs.head()
```

```python
town_runs.to_json("../data/town_runs.json",orient="index")
```
